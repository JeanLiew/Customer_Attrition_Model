```python
#import pandas 

import pandas as pd
pd.set_option('display.max_columns',100)
```

## Import Datasets 

```python
#demographics
demo = pd.read_excel('../data/Telco_customer_churn_demographics.xlsx')
#customer churn 
churn = pd.read_excel('../data/Telco_customer_churn.xlsx')
#services provided
service = pd.read_excel('../data/Telco_customer_churn_services.xlsx')
#status of customer
status = pd.read_excel('../data/Telco_customer_churn_status.xlsx')
#pop
# population = pd.read_excel('../data/Telco_customer_churn_population.xlsx')
#pop_density
pop = pd.read_csv('../data/pop_ca_density.csv')
```

```python
demo.head(1)
```

```python
churn.head(1)
```

```python
#rename Customer ID name for consistency
churn = churn.rename(columns={'CustomerID':'Customer ID'})
```

```python
service.head(1)
```

```python
status.head(1)
```

```python
#merging df together 
df = pd.merge(left=demo,right=churn, on='Customer ID', suffixes=[None, '_y'])
df = pd.merge(left=df,right=service, on='Customer ID', suffixes=[None, '_y'])
df = pd.merge(left=df,right=status, on='Customer ID', suffixes=[None, '_y'])

df.columns = df.columns.str.lower().str.replace(' ','_')
df.head()
```

```python
df.columns.sort_values()
```

```python
#identify common columns
common_col = [col for col in df.columns if col.endswith('_y')]
common_col
```

```python
#checking similar columns on merged df to verify before drop
check_col =[]
drop_col = ['count','count_y']
for i in range(1,len(common_col)):
    compare_col = common_col[i].replace('_y','')
    #exempt Count_y as there are several 
    #all count_y shall be dropped
    if common_col[i] == 'count_y':
        pass
    elif ((df[compare_col] != df[common_col[i]]).sum()) != 0:
        check_col.append(common_col[i])
    else:
        drop_col.append(common_col[i])
        

print(f'Columns to check: {check_col}')
print('\n')
print(f'Columns to drop: {drop_col}')
```

```python
#investigating columns that is not consistent
df[(df['total_charges'] != df['total_charges_y'])][['total_charges','total_charges_y']]
```

```python
df[(df['churn_score'] != df['churn_score_y'])][['churn_score','churn_score_y']]
```

```python
df[(df['churn_reason'] != df['churn_reason_y'])][['churn_reason','churn_reason_y']]
```

```python
df[(df['monthly_charge'] == df['monthly_charges'])][['monthly_charge','monthly_charges']]
```

```python
#columns to drop
df = df.drop(columns=drop_col)

#additional columsn to drop after manual checks
final_drop_col = ['multiple_lines_y', 'internet_service_y', 'online_security_y', 'online_backup_y', 'streaming_tv_y', 'streaming_movies_y', 'contract_y', 
                  'payment_method_y', 'total_charges', 'churn_score_y', 'churn_reason_y', 'monthly_charges']

df = df.drop(columns=final_drop_col)
```

```python
df.columns.sort_values()
```

```python
#cleaning pop dataset
pop.head()
```

```python
#adding population count
pop = pop.drop(columns=['land_area','population'])
```

```python
#merge pop count on df
df = pd.merge(df, pop, on='zip_code',how='left')
```
