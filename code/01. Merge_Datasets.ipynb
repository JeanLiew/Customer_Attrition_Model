```python
import pandas as pd
pd.set_option('display.max_columns',100)
```

## Import Datasets 

```python
#demographics
demo = pd.read_excel('../data/Telco_customer_churn_demographics.xlsx')
demo = demo.set_index('Customer ID')
#customer churn 
churn = pd.read_excel('../data/Telco_customer_churn.xlsx')
churn = churn.set_index('CustomerID')
#services provided
service = pd.read_excel('../data/Telco_customer_churn_services.xlsx')
service = service.set_index('Customer ID')
#status of customer
status = pd.read_excel('../data/Telco_customer_churn_status.xlsx')
status = status.set_index('Customer ID')
# population
pop = pd.read_excel('../data/Telco_customer_churn_population.xlsx')
```

```python
df = pd.concat([demo,churn,service,status],axis=1).reset_index()
```

```python
df.info()
```

```python
#flip duplicate boolean and keep the first
df = df.loc[:,~df.columns.duplicated()]

df.columns = df.columns.str.lower().str.replace(' ','_')
df.columns
```

```python
df.head()
```

```python
#adding population count
pop = pop.rename(columns={'Zip Code':'zip_code', 'Population':'population'}).drop(columns=['ID'])
```

```python
df = pd.merge(df, pop, on='zip_code')
```

## Export Dataset

```python
df.to_csv('../data/telco_churn_data',index=False)
```
