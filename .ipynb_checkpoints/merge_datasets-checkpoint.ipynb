```python
import pandas as pd
pd.set_option('display.max_columns',100)
```

## Import Datasets 

```python
#demographics
demo = pd.read_excel('./Telco_customer_churn_demographics.xlsx')
demo = demo.set_index('Customer ID')

#customer churn 
churn = pd.read_excel('./Telco_customer_churn.xlsx')
churn = churn.set_index('CustomerID')

#services provided
service = pd.read_excel('./Telco_customer_churn_services.xlsx')
service = service.set_index('Customer ID')

#status of customer
status = pd.read_excel('./Telco_customer_churn_status.xlsx')
status = status.set_index('Customer ID')

#location of customer base
loc = pd.read_excel('./Telco_customer_churn_location.xlsx')
loc = loc.set_index('Customer ID') 

# population
pop = pd.read_excel('./Telco_customer_churn_population.xlsx')
```

```python
#concat data sets except population data
df = pd.concat([demo, churn, service, status, loc], axis=1)
```

```python
df.info()
```

```python
#flip duplicate boolean and keep the first
df = df.loc[:,~df.columns.duplicated()]
```

```python
df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
```
